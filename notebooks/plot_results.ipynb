```python
import torch
from torch.nn import functional as F
```

```python
x = torch.tensor([[1., 3.], [4., 2.]])
y = torch.tensor([[2., 4.], [4., 1.]])
```

```python
torch.mean(torch.sum(- F.softmax(y / 2, dim=1) * F.log_softmax(x / 2, dim=1), 1))
```

```python
F.binary_cross_entropy(F.softmax(x / 2, dim=1), F.softmax(y / 2, dim=1))
```
